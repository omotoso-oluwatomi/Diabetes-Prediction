```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm #svm's the Support vector machine.
from sklearn.metrics import accuracy_score
from streamlit_jupyter import StreamlitPatcher
StreamlitPatcher().jupyter()
import streamlit as st
import joblib
import logging
logging.getLogger("streamlit").setLevel(logging.ERROR)
```

Import the data.

```python
filePath = "./diabetes.csv"
diabetes_dataset = pd.read_csv(filePath)
diabetes_dataset.head()
#From our dataset we can see that our feature set all combine to infer whether the patient is diabetic or not [1 or 0].

 
```

DataSets Stats

```python
diabetes_dataset.shape
#We have 768 columns so we're going to use only the first 767 columns
```

```python
#Statistical measures
diabetes_dataset.describe()
```

```python
diabetes_dataset[["Outcome"]].value_counts()
```

1 [Diabetic]
0[Non Diabetic]

```python
diabetes_dataset.groupby('Outcome').mean()
```

Seperating the labels from the data

```python
data = diabetes_dataset.drop(columns = 'Outcome', axis = 1) 
# we say axis = 1 if it's a column and axis = 0 if it's a row
labels = diabetes_dataset['Outcome']
```

Data Standardization

```python
scaler = StandardScaler()
scaler.fit(data)
standardized_data = scaler.transform(data)
```

```python
data = standardized_data
#Label's still the outcome column.
```

```python
print(data)
```

Splitting the data into test and train_test_split

```python
data_train, data_test, labels_train, labels_test = train_test_split(data,labels,test_size= 0.2, stratify= labels, random_state= 2)
```

```python
#Debugging
print(data.shape,data_train.shape,data_test.shape)
```

Training the data

```python
classifier = svm.SVC(kernel = 'linear')
```

Training the SVM classifier

```python
classifier.fit(data_train,labels_train)
```

Model Evaluation

```python
#Accuracy Score --> Trainig data
data_train_prediction = classifier.predict(data_train)
training_data_accuracy = accuracy_score(data_train_prediction,labels_train)
```

```python
print("Acuuracy score of trainig data: ",training_data_accuracy*100)
#This is good but it can be better,
```

```python
#Accuracy Score --->Testing data
data_test_prediction = classifier.predict(data_test)
testing_data_accuracy= accuracy_score(data_test_prediction,labels_test)
```

```python
print("Accuracy score of testing data: ",testing_data_accuracy*100)
```

Making a Predictive system

Getting Inputs

```python
data_for_model_evaluation = []
gender =  input("Are you male of female? ")

#Pregnancies
if(gender.lower()[0] == 'm'):
    preg = 0
else:
    preg = int(input("How many times have you been pregnant: "))

data_for_model_evaluation.append(preg)


#Glucose level
gluc = int(input("What's your glucose level: "))
data_for_model_evaluation.append(gluc)

#BP
bp = int(input("What's your Blood Pressure"))
data_for_model_evaluation.append(bp)

#Sking Thickness
skin_thick = int(input("What's your skin thickness"))
data_for_model_evaluation.append(skin_thick)

#Insulin
insulin = int(input("What's your insulin level"))
data_for_model_evaluation.append(insulin)

#BMI
bmi = float(input("What's your BMI level: "))
data_for_model_evaluation.append(bmi)

#DiabetesPedigreeFuncton
DPF = float(input("What's your Diabetes Pedigree Function: "))
data_for_model_evaluation.append(DPF)

#Age
age = int(input("And how old are you: "))
data_for_model_evaluation.append(age)

```

Prediction

```python
np_data_for_model_evaluation = np.asarray(data_for_model_evaluation)

reshaped_np_data_for_model_evaluation = np_data_for_model_evaluation.reshape(1,-1)

std_data = scaler.transform(reshaped_np_data_for_model_evaluation)

prediction = classifier.predict(std_data)
if prediction[0]== 1:
    print("Oooh you've got a huge chance at being diabetic")
elif prediction[0]==0:
    print("You're alright, there's little chance you're diabetic.")

```
